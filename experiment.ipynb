```python
import os
import re
import error_calculator
import my_segment
import my_func
```

```python
list_en = os.listdir(path="./en") 
```

```python
count_of_files = 1 # количество факлов для тестирования
```

```python
def test_(name_alg):
    total_tp = 0
    total_fp = 0
    total_fn = 0
    for i in list_en[:count_of_files]:
        print(i)
        f = open("./en/" + i, 'r', encoding="utf8")
        text = f.read()
        text = re.sub('<[^.]+>', '', text).replace('\n', ' ')
        len_part_text = 5000
        count = len(text) // len_part_text
        text_main = my_func.my_processing(text)

        for j in range(count):
            print(j, '/', count)
            text_part = text_main[j * len_part_text: (j + 1) * len_part_text]
            my_func.my_save_file('text_main.txt', text_part)
            text_without_punct = text_part.replace('.', '').lower()
            text_punctuator = my_segment.segment(text_without_punct, name_alg)
            my_func.my_save_file(name_alg+'.txt', text_punctuator)
            try : 
                tp, fp, fn = error_calculator.compute_error(['text_main.txt'], [name_alg+'.txt'])
                total_tp += tp
                total_fp += fp
                total_fn += fn
            except: continue
    pre = total_tp/(total_tp+total_fp) if total_fp else nan
    rec = total_tp/(total_tp+total_fn) if total_fn else nan
    f1 = (2.*pre*rec)/(pre+rec) if (pre + rec) else nan
    print(name_alg,': ', pre, rec, f1)
    return(pre, rec, f1)
```

```python
NNsplit_pre, NNsplit_rec, NNsplit_f1 = test_('NNsplit')
```

```python
punctuator_pre, punctuator_rec, punctuator_f1 = test_('punctuator')
```
